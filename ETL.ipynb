### MERGED_DF

```python
# Realiza un merge de ambos DataFrames 
merged_df = pd.merge(df_hechos, df_vic, left_on='ID', right_on='ID_hecho')


```

```python
merged_df.isnull().sum()
```

```python
merged_df.reset_index(drop=True)
```

```python
#Borramos los NaN 

merged_df = merged_df.dropna()
```

```python
merged_df
```

```python
#Borramos la columna ID_hecho

merged_df = merged_df.drop(['ID_hecho'], axis=1)
```

```python
#Borramos las columnas AAAA_y y MM_y y DD_y

merged_df = merged_df.drop(['AAAA_y', 'MM_y', 'DD_y'], axis=1)
```

```python
#Renombramos las columnas AAAA_x, MM_x y DD_x a AAAA, MM y DD

merged_df = merged_df.rename(columns={'AAAA_x': 'AAAA', 'MM_x': 'MM', 'DD_x': 'DD'})
```

```python
#Convertimos a int las columnas AAAA, MM , DD , HH y COMUNA

merged_df['AAAA'] = merged_df['AAAA'].astype(int)
merged_df['MM'] = merged_df['MM'].astype(int)
merged_df['DD'] = merged_df['DD'].astype(int)
merged_df['HH'] = merged_df['HH'].astype(int)
merged_df['COMUNA'] = merged_df['COMUNA'].astype(int)
```

```python
#Lectura del dataframe

merged_df
```

```python
#Borramos la columna Hora

merged_df = merged_df.drop(['Hora'], axis=1)
```

```python
#Lectura del dataframe

merged_df
```

```python
#Borramos la columna FECHA_y

merged_df = merged_df.drop(['FECHA_y'], axis=1)
```

```python
#Cambiamos los tipos de datos de la columna EDAD a int

merged_df['EDAD'] = merged_df['EDAD'].astype(int)
```

```python
#Cambiamos los tipos de datos de la columna  N_VICTIMAS a int

merged_df['N_VICTIMAS'] = merged_df['N_VICTIMAS'].astype(int)
```

```python
#Lectura del dataframe

merged_df
```

```python
#Borramos la columna ID

merged_df = merged_df.drop(['ID'], axis=1)
```

```python
#Reseteamos el index nuevamente

merged_df = merged_df.reset_index(drop=True)
```

```python
#Lectura del dataframe

merged_df
```

```python
#Guardamos el dataframe en un archivo csv

merged_df.to_csv('merged_df.csv', index=False)
```
